The technology categories with 0 have no recorded electricity generation values in the dataset, likely due to aggregation, classification, or reporting limitations. The dataset mixes actual technologies (usable for analysis) and aggregate categories (should be filtered out). 
1.“Solar energy” is just a parent category, not a real measured one.
2.Multiple renewables* is probably combined/mixed sources and is not individually tracked. It is usually excluded in detailed analysis.
3.Other renewable energy is a vague category.It is often rarely reported and inconsistently used.
4.Non-renewable municipal waste can generate electricity, but may not be reported in the dataset.

```python
invalid_tech = [
    "Solar energy",
    "Multiple renewables*",
    "Other renewable energy"
]

noinvalidtech_country_df = country_df[~country_df["Technology"].isin(invalid_tech)]
```

```python
noinvalidtech_country_df.isna().sum()
```

```python
noinvalidtech_country_df.shape
```

For Electricity Generation Analysis
The dataset still has missing values because; 1. Some countries didn’t report data, 2.Some technologies exist but had no production that year, 3. Some metrics don’t apply in certain cases

```python
generation_df = noinvalidtech_country_df.dropna(subset=["Electricity Generation (GWh)"])
```

```python
generation_df.isna().sum()
```

```python
generation_df.shape
```

Do not handle outliers because in energy data:
Countries like China, USA, Brazil produce HUGE amounts of electricity
Small countries produce very little
Those are not errors — they are real-world differences
So whhat looks like an “outlier” is often actually the most important insight

```python
generation_df.describe()
```

The minimum value for Electricity Generation (the negative) is not realistic in this context. So all the negatives have to be removed. It could be it could be a data entry error or accounting adjustments (rare, but not useful for your analysis). The very high values are important in this context so should be maintained. The columns that show 0 count still exist but are entirely NaN in this filtered dataset. And so for this analysis we shall keep only the relevant columns.

```python
generation_df = generation_df[generation_df["Electricity Generation (GWh)"] >= 0]
```

```python
generation_df.describe()
```

```python
generation_df = generation_df[[
    "Country",
    "Region",
    "Technology",
    "Year",
    "Electricity Generation (GWh)"
]]
```

```python
#Generation Over Time
gen_year = generation_df.groupby("Year")["Electricity Generation (GWh)"].sum().reset_index()
```

```python
print(gen_year)
```

```python
#Generation by Technology
gen_tech = generation_df.groupby("Technology")["Electricity Generation (GWh)"].sum().reset_index()
```

```python
print(gen_tech)
```

```python
#Generation by Region
gen_region = generation_df.groupby("Region")["Electricity Generation (GWh)"].sum().reset_index()
```

```python
print(gen_region)
```

```python
#Top Countries
top_countries = (
    generation_df.groupby("Country")["Electricity Generation (GWh)"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)
```

```python
print(top_countries)
```
